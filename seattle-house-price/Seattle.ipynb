```python
def myscaler(arr,col):
    ans = ((arr[col] - arr[col].min())/arr[col].max()).values
    return ans

def SeperateData(df,cols):
    new_df = df
    for col in cols:
        new_df[col] = pd.Series(data=myscaler(new_df,col),name=col)
    return new_df,cols

def makeML(df,x_cols,y_col,zipcode,zipcodes):
    # get train and test split
    all_cols = x_cols + y_col
    sub_df = df.loc[df["zipcode"]==zipcode,all_cols]
    x_train, x_test, y_train, y_test = train_test_split(sub_df[x_cols],sub_df[y_col],test_size=0.3)
    tree_model = RandomForestClassifier(max_depth=4,criterion='gini')
    tree_model.fit(x_train,y_train)

    results = []
    for zippy in zipcodes:
        if zippy == zipcode:
            results.append(mean_squared_error(tree_model.predict(x_test),y_test))
        else:
            y_pred = tree_model.predict(df.loc[df["zipcode"]==zippy,x_cols])
            results.append(tree_model.score(y_pred,df.loc[df["zipcode"]==zippy,y_col].values))
    
    return pd.Series(data=results,name=zipcode)
        
        
def makeML_master(df,x_cols,y_col,zipcodes):
    # preprocess the data

    p = ThreadPool(7)
    ##df.dropna(inplace=True)
    
    df,cols = SeperateData(df,x_cols)
    
    ##df.dropna(inplace=True)
    
    # map the list of zipcodes to the ML function
    results = p.map(lambda x: makeML(df,x_cols,y_col,x,zipcodes),zipcodes)
    results = pd.DataFrame(results)
    return (results,df)
```

```python
def plotAB(df,a_col):
    b_cols = [x for x in df.columns if x != a_col]
    b_cols = [x for x in b_cols if x not in ["id","date"]]
    
    figs, axes = plt.subplots(nrows=6,ncols=3,figsize=(15,30))
    cheap_fix = fixPlot(axes)

    for b_col in b_cols:
        df.plot(kind="scatter",y=a_col,x=b_col,ax=cheap_fix.pop(0),alpha=0.5,color='r')
        #ax.scatter(df[a_col],df[b_col],ax=ax)
    figs.tight_layout()
    plt.show()
        
    
```

```python
def fixPlot(axes):
    figures = []
    for i in axes:
        for j in i:
            figures.append(j)
    return figures
        
```
