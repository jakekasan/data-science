# 2015 Airport Delays
## Exploratory data analysis

### Synopsis

This EDA looks into flight delay data for the United States for the year 2015. The objective here is to provide as much insight as possible into the domestic flights market as possible, speficically the differences in a key performance metric, flight delays, between the airlines


The dataset comes in three parts: Airport data, Flights data and Airline data. The flights data contains the origin and destination airports, scheduled and actual departure times, arrival times, flight times as well as the airline and flight number of that plane.


### Analysis

#### Intro
The dataset covers flights scheduled to take off from January 1st, 2015 to December 31st, 2015 within United States, including Alaska, Hawaii and Puerto Rico. 

#### Initial Notes

Before beginning the EDA, it is important to note a few things:

First, this dataset includes cancelled flights. For these, variables with record actual events (as opposed to scheduled events) such as "WHEELS_OFF" or "ARRIVAL_DELAY" are NaN. While these flights could be dropped altogether of have the NaNs filled with zero or the mean, this would unnecessarily distort the dataset and inferernces. The cancelled flights are indicative of airline performance, so dropping them from consideration altogether leaves an incomplete picture, while substituting the NaNs for the zero or mean adds bias to the dataset. To resolve this, the cancelled flights are seperated into a different dataset and examined individually.

```python
airports = pd.read_csv("../data/flight-delays/airports.csv")
airlines = pd.read_csv("../data/flight-delays/airlines.csv")
flights = pd.read_csv("../data/flight-delays/flights.csv")
```

## Data Cleaning

### NaN Values

As stated before, to drop the NaNs altogether or to substitute them with another value detracts from the analysis, so the cancelled and diverted flights are spun off into a seperate data frame, while the NaNs in the specific delays columns ("AIR_SYSTEM_DELAY", "SECURITY_DELAY", etc...) are filled in with 0, as a NaN in that field signifies that there was no delay due to that specific cause or that the flight was cancelled. 

```python

binaries = ["CANCELLATION_REASON","AIR_SYSTEM_DELAY","SECURITY_DELAY","AIRLINE_DELAY","LATE_AIRCRAFT_DELAY","WEATHER_DELAY"]
flights[binaries] = flights[binaries].fillna(value=0)
flights.dropna()


cancelled_flights = flights.loc[(flights["CANCELLED"] == 1) & (flights["DIVERTED"] != 1) ,:]
diverted_flights = flights.loc[(flights["CANCELLED"] != 1) & (flights["DIVERTED"] == 1) ,:]
flights = flights.loc[(flights["CANCELLED"] != 1) & (flights["DIVERTED"] != 1) ,:]
print("Null check... \n")
for column in flights.columns:
    print("{}: {}".format(column,flights[column].isnull().any()))

print("\n\nSizes: \nCancelled: {}\nDiverted: {}\nOther: {}\n".format(len(cancelled_flights.index),len(diverted_flights.index),len(flights.index)))
```

```python
# Get the Mean, std Dev, and 25%-75% quantiles of the delays per airline

airline_stats = airlines.set_index("IATA_CODE").T.to_dict()


for key in airline_stats.keys():
    departure_delays = flights.loc[(flights["AIRLINE"] == key) , "DEPARTURE_DELAY"]
    arrival_delays = flights.loc[(flights["AIRLINE"] == key) , "ARRIVAL_DELAY"]
    cancelled = cancelled_flights.loc[(cancelled_flights["AIRLINE"] == key),"AIRLINE"]
    diverted = diverted_flights.loc[(diverted_flights["AIRLINE"] == key),"AIRLINE"]
    
    airline_stats[key]["departure_delay_average"] = np.mean(departure_delays)
    airline_stats[key]["arrival_delay_average"] = np.mean(arrival_delays)
    airline_stats[key]["departure_delay_stdev"] = np.std(departure_delays)
    airline_stats[key]["arrival_delay_stdev"] = np.std(arrival_delays)
    airline_stats[key]["departure_delay_25per"] = np.percentile(departure_delays,25)
    airline_stats[key]["arrival_delay_25per"] = np.percentile(arrival_delays,25)
    airline_stats[key]["departure_delay_75per"] = np.percentile(departure_delays,75)
    airline_stats[key]["arrival_delay_75per"] = np.percentile(arrival_delays,75)
    airline_stats[key]["arrival_delayed_under_5_mins"] = arrival_delays.apply(lambda x: x < 5).sum()
    airline_stats[key]["arrival_delayed_under_30_mins"] = arrival_delays.apply(lambda x: x < 30).sum()
    airline_stats[key]["arrival_delayed_under_1_hour"] = arrival_delays.apply(lambda x: x < 60).sum()
    airline_stats[key]["arrival_delayed_under_2_hours"] = arrival_delays.apply(lambda x: x < 120).sum()
    airline_stats[key]["arrival_delayed_over_2_hours"] = arrival_delays.apply(lambda x: x >= 120).sum()
    airline_stats[key]["departure_delayed_under_5_mins"] = departure_delays.apply(lambda x: x < 5).sum()
    airline_stats[key]["departure_delayed_under_30_mins"] = departure_delays.apply(lambda x: x < 30).sum()
    airline_stats[key]["departure_delayed_under_1_hour"] = departure_delays.apply(lambda x: x < 60).sum()
    airline_stats[key]["departure_delayed_under_2_hours"] = departure_delays.apply(lambda x: x < 120).sum()
    airline_stats[key]["departure_delayed_over_2_hours"] = departure_delays.apply(lambda x: x >= 120).sum()
    
    # now percentages...
    airline_stats[key]["arrival_delayed_under_5_mins_percent"] = arrival_delays.apply(lambda x: x < 5).sum()/arrival_delays.count()
    airline_stats[key]["arrival_delayed_under_30_mins_percent"] = arrival_delays.apply(lambda x: x < 30).sum()/arrival_delays.count()
    airline_stats[key]["arrival_delayed_under_1_hour_percent"] = arrival_delays.apply(lambda x: x < 60).sum()/arrival_delays.count()
    airline_stats[key]["arrival_delayed_under_2_hours_percent"] = arrival_delays.apply(lambda x: x < 120).sum()/arrival_delays.count()
    airline_stats[key]["arrival_delayed_over_2_hours_percent"] = arrival_delays.apply(lambda x: x >= 120).sum()/arrival_delays.count()
    airline_stats[key]["departure_delayed_under_5_mins_percent"] = departure_delays.apply(lambda x: x < 5).sum()/departure_delays.count()
    airline_stats[key]["departure_delayed_under_30_mins_percent"] = departure_delays.apply(lambda x: x < 30).sum()/departure_delays.count()
    airline_stats[key]["departure_delayed_under_1_hour_percent"] = departure_delays.apply(lambda x: x < 60).sum()/departure_delays.count()
    airline_stats[key]["departure_delayed_under_2_hours_percent"] = departure_delays.apply(lambda x: x < 120).sum()/departure_delays.count()
    airline_stats[key]["departure_delayed_over_2_hours_percent"] = departure_delays.apply(lambda x: x >= 120).sum()/departure_delays.count()
    airline_stats[key]["cancelled_rate"] = cancelled.count() / (cancelled.count() + diverted.count() + arrival_delays.count())
    airline_stats[key]["diverted_rate"] = diverted.count() / (cancelled.count() + diverted.count() + arrival_delays.count())
    
airline_stats = pd.DataFrame.from_dict(airline_stats,orient="index")
```

```python
# Get the Mean, std Dev, and 25%-75% quantiles of the delays per airline

def getStats(airlines,delays=[5,30,60,120]):
    airline_stats = airlines.set_index("IATA_CODE").T.to_dict()
    
    for key in airline_stats.keys():
        total = flights.loc[(flights["AIRLINE"] == key) , ["DEPARTURE_DELAY","ARRIVAL_DELAY"]]
        cancelled = cancelled_flights.loc[(cancelled_flights["AIRLINE"] == key),"AIRLINE"]
        diverted = diverted_flights.loc[(diverted_flights["AIRLINE"] == key),"AIRLINE"]
        
        flight_count = cancelled.count() + diverted.count() + total["ARRIVAL_DELAY"].count()
        
        last_delay = 0
        for delay in delays:
            if last_delay != 0:
                airline_stats[key]["arrival_{}min_to_{}min".format(last_delay,delay)] = total["ARRIVAL_DELAY"].apply(lambda x: (x > last_delay) & (x <= delay)).sum() / total["ARRIVAL_DELAY"].count()
            else:
                airline_stats[key]["arrival_to_{}min".format(delay)] = total["ARRIVAL_DELAY"].apply(lambda x: (x <= delay)).sum() / total["ARRIVAL_DELAY"].count()
            last_delay = delay
            
        airline_stats[key]["arrival_to_{}min".format(delay)] = total["ARRIVAL_DELAY"].apply(lambda x: (x > last_delay)).sum() / total["ARRIVAL_DELAY"].count()

    airline_stats = pd.DataFrame.from_dict(airline_stats,orient="index")
    
    return airline_stats

airstat = getStats(airlines,delays=[5,30,60,120])
```

```python
def plotBars(airlines,columns):
    fig, ax = plt.subplots(ncols=1,nrows=1,figsize=(15,15))
    bottom = [0 for x in airlines.index]
    plots = []
    for column in columns:
        plots.append(ax.barh(np.arange(airlines[column].count()),airlines[column],left=bottom))
        bottom = [a+b for a,b in zip(bottom,airlines[column])]

    plt.setp(ax,yticks=np.arange(len(airlines.index)),yticklabels=airlines["AIRLINE"])
    #ax.xticks(np.arange(len(airlines.index)),airlines.index)
    
    box = ax.get_position()
    ax.set_position([box.x0,box.y0 + box.height * 0.1, box.width, box.height * 0.9])
    legend_labs = ["Delayed less than 5 minutes","Delayed between 5 and 30 minutes","Delayed between 30 and 60 minutes","Delayed between 60 and 120 minutes","Delayed over 120 minutes"]
    ax.legend([x[0] for x in plots],legend_labs,loc="upper center",fancybox=True,shadow=True,bbox_to_anchor=(0.5,-0.05),ncol=3)
    ax.set_title("Breakdown of Delay Percentages")
    
    plt.show()
    
plotBars(airstat,[x for x in airstat.columns if "arrival" in str(x)])
#print([x for x in airstat.columns if "arrival" in str(x)])
#print(flights.loc[flights["AIRLINE"] == "AS","ARRIVAL_DELAY"].apply(lambda x: (x > 400) & (x > 500)).sum())
```

```python
flights.loc[(flights["DESTINATION_AIRPORT"] == "ABR") | (flights["ORIGIN_AIRPORT"] == "ABR"),["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]].drop_duplicates()

def getConnections(row):
    query = flights.loc[(flights["DESTINATION_AIRPORT"] == row["IATA_CODE"]) | (flights["ORIGIN_AIRPORT"] == row["IATA_CODE"]),["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]]
    #print("Number of Connections = {}".format(query.drop_duplicates().shape[0]))
    #print("Number of flights per day = {}".format(query.shape[0]/365))
    row["connections"] = query.drop_duplicates().shape[0]
    row["flights_per_day"] = query.shape[0]/365
    return row
    

airports = airports.apply(getConnections,axis=1)
```

```python
# one hot encoding for binaries other than days of the week

def getBinaries(binaries,df):
    for binary in binaries:
        elements = [x for x in df[binary].unique() if x in airports["IATA_CODE"].unique() or x in airlines["IATA_CODE"].unique()]
        for element in elements:
            df["{}_{}_binary".format(binary,element)] = df[binary].apply(lambda x: 1 if x == element else 0).astype(np.int8)
        print("Completed {}".format(binary))
    return df

def getWeekBinaries(binary,df,dow_key):
    for element in df[binary].unique():
        for key, value in dow_key.items():
            if value == element:
                df["{}_{}_binary".format(binary,key)] = df[binary].apply(lambda x: 1 if x == element else 0).astype(np.int8)
                return df
```

```python
def countUniques(flights,row,target):
    vals = target.values()
    keys = target.keys()
    if 0 not in vals:
        return
    if (row["DESTINATION_AIRPORT"] in vals) & (row["ORIGIN_AIRPORT"] in vals):
        return
    if (row["DESTINATION_AIRPORT"] in vals) & (row["ORIGIN_AIRPORT"] not in vals) & (row["ORIGIN_AIRPORT"] not in keys):
        query = flights.loc[ (np.absolute(flights["SCHEDULED_DEPARTURE"] - row["SCHEDULED_DEPARTURE"]) < 1000) & ((flights["DISTANCE"] - row["DISTANCE"]) < 50) & (flights["AIRLINE"] == row["AIRLINE"]) & (flights["MONTH"] != row["MONTH"]), ["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]].drop_duplicates()
        if query.shape[0] == 1:
            target[query.values[0][1]] = str(row["ORIGIN_AIRPORT"])
        return
    if (row["DESTINATION_AIRPORT"] not in vals) & (row["DESTINATION_AIRPORT"] not in keys) & (row["ORIGIN_AIRPORT"] in vals):
        query = flights.loc[ (np.absolute(flights["SCHEDULED_DEPARTURE"] - row["SCHEDULED_DEPARTURE"]) < 1000) & ((flights["DISTANCE"] - row["DISTANCE"]) < 50) & (flights["AIRLINE"] == row["AIRLINE"]) & (flights["MONTH"] != row["MONTH"]), ["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]].drop_duplicates()
        if query.shape[0] == 1:
            target[query.values[0][1]] = str(row["DESTINATION_AIRPORT"])
        return
    elif ((row["DESTINATION_AIRPORT"] not in keys) & (row["ORIGIN_AIRPORT"] not in keys) & (row["DESTINATION_AIRPORT"] not in vals) & (row["ORIGIN_AIRPORT"] not in vals)):
        query = flights.loc[ (np.absolute(flights["SCHEDULED_DEPARTURE"] - row["SCHEDULED_DEPARTURE"]) < 1000) & ((flights["DISTANCE"] - row["DISTANCE"]) < 50) & (flights["AIRLINE"] == row["AIRLINE"]) & (flights["MONTH"] != row["MONTH"]), ["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]].drop_duplicates()
        if query.shape[0] == 1:
            target[query.values[0][0]] = str(row["DESTINATION_AIRPORT"])
            target[query.values[0][1]] = str(row["ORIGIN_AIRPORT"])
        return
    
    

#target = {x:0 for x in airports["IATA_CODE"].unique()}
#print(target)
#for airline in target:
    
october = flights.loc[(flights["MONTH"] == 10), :]
#sub_october = october.loc[(~flights["ORIGIN_AIRPORT"].isin(target.values())) & (~flights["DESTINATION_AIRPORT"].isin(target.values())) & (flights["DAY"] < 3) & (flights["DAY"] > 1)  , :]
sub_october = october.loc[ (~october["DESTINATION_AIRPORT"].isin(target.values())) & (~october["ORIGIN_AIRPORT"].isin(target.values())) & (october["DAY"] > 0) & (october["DAY"] < 34),]
sub_october.apply(lambda x: countUniques(flights,x,target),axis=1)
#print(target)
#sub_october.loc[:,["DESTINATION_AIRPORT","ORIGIN_AIRPORT"]].drop_duplicates()
```
